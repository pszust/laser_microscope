```python
wd = 'c:/Users/LAB218/Documents/mikroskop2/'
def read_script(fname, arguments = []):
    def clean_cmd(cmd):
        cmd = cmd[:cmd.find('#')]
        clean = cmd[:cmd.find('(')]
        args = cmd[cmd.find('(')+1:cmd.find(')')]
#         args = args.replace(',', ' ')
        while args.find(' ') != -1:
            args = args.replace(' ', '')
        return [clean, args.split(',')]
    
    possibleLocations = [fname, wd + fname, wd + 'scripts/' + fname, wd + 'scripts/base/' + fname]
    path = ''
    for loc in possibleLocations:
        print(loc + '.txt')
        if os.path.isfile(loc + '.txt') == True:
            path = loc + '.txt'
            break
    
    if path == '':
        print('Script %s not found'%fname)
        return -1
    
    with open(path, 'r') as f:
        cnt = f.read()
    
    for i, arg in enumerate(arguments):
        cnt = cnt.replace('arg' + str(i).zfill(2), arg)
    
    if cnt.find('arg') >= 0:
        print('Not enough arguments for script %s'%fname)
        return -2
    
    cnt = cnt.split('\n')
    
    answer = []
    for cmd in cnt:
        if cmd.find('(') >= 0:
            ans = clean_cmd(cmd)
            if ans[0] != '':
                answer.append(ans)
    return answer
        
    
sc = read_script('test_script')
sc
```

```python
class Proc:
    def __init__(self):
        self.scripts = []  # holds lists of commands to be executed
        self.lines = []  # keeps track of commands execution        
        self.ctime = 0
        
        self.scripts.append(read_script('test_script2.txt'))
        self.lines = [0]
        
    def main_loop(self):
        if len(self.scripts) > 0:
            comNum = self.lines[-1]
            if comNum == len(self.scripts[-1]):
                print('script %d finished'%len(self.scripts))
                del self.lines[-1]
                del self.scripts[-1]
            else:
                cmd = self.scripts[-1][comNum]
                self.lines[-1] += 1
                self.execute_command(cmd)
        else:
            print('No script to execute')
            
    def execute_command(self, command):
        cmd, args = command[0], command[1]        
        print('executing', cmd, args)
        
        if cmd == 'd_wait':
            if time.time() < self.ctime:
                self.lines[-1] -= 1             
            if self.ctime == 0:
                self.ctime = time.time() + float(args[0])
                self.lines[-1] -= 1        
                print('wait started for %2.1f'%float(args[0]))
            else:
                if time.time() > self.ctime:
                    self.ctime = 0
                    print('wait finished')
        
        if cmd[:2] != 'd_':
            self.scripts.append(read_script('scripts/%s.txt'%cmd, arguments = args))
            self.lines.append(0)
            _ = [print(sc) for sc in self.scripts]
            print(self.lines)
        
        print('')
        
        
proc = Proc()
```

```python
for i in range(0, 40):
    proc.main_loop()
    time.sleep(0.5)
```

# TC300

```python
try:
    from TC300_COMMAND_LIB import *
    import time
except OSError as ex:
    print("Warning:",ex)
         
```

```python
devs = TC300ListDevices()
print(devs)
if(len(devs)<=0):
    print('There is no devices connected')
    exit()     
```

```python
devs[0][1]
```

```python
TC300= devs[0]
serialNumber= TC300[0]
hdl = TC300Open(serialNumber,115200,3)       
if(hdl < 0):
    print("Connect ",serialNumber, "failed" )
else:
    print("Connect ",serialNumber, "successfully")
```

```python
TC300IsOpen(serialNumber)
```

```python
TC300EnableChannel(hdl, 1, 1)
```

```python
mode=[0]
modeList={0: "Heater", 1: "Tec", 2: "Constant current", 3: "Synchronize with Ch1(only for channel 2)"}
result=TC300GetMode(hdl,1,mode)
if(result<0):
    print("Get mode fail",result)
else:
    print("Get mode :",modeList.get(mode[0]))
```

```python
TargetTemperature=[0]
result=TC300GetTargetTemperature(hdl,1,TargetTemperature)
if(result<0):
    print("Get Target Temperature fail",result)
else:
    print("Target Temperature is :",TargetTemperature[0])
```

```python
ActualTemperature=[0]
result=TC300GetActualTemperature(hdl,1,ActualTemperature)
if(result<0):
    print("Get Actual Temperature fail",result)
else:
    print("Actual Temperature is :",ActualTemperature[0]) 
```

```python
TC300SetTargetTemperature(hdl, 1, 50)
```
